```python
# Correlation Analysis for Feature Selection

# Exclude columns that are object types
numerical_features = main_merged_dataset.select_dtypes(include=['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_features.corr()

# Set the threshold for high correlation
# Based on domain knowledge, we will set a threshold of 0.85
# Features with correlation above this threshold will be removed to avoid multicollinearity
correlation_threshold = 0.85

# Identify highly correlated features
highly_correlated_features = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > correlation_threshold:
            colname = correlation_matrix.columns[i]
            highly_correlated_features.append(colname)

# Drop the highly correlated features from the dataset
features_corr_reduced = numerical_features.drop(columns=highly_correlated_features)

# Print the reduced features
print("Reduced features shape:", features_corr_reduced.shape)

# Reasoning:
# - High correlation between features indicates redundancy, meaning the features are providing similar information.
# - Removing these features helps in reducing multicollinearity, which can negatively impact model interpretability and stability.
# - This step ensures that the model is trained on unique, informative features, improving overall generalizability.
```

```python
# Comprehensive Feature Selection and Elimination

from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_regression
from sklearn.feature_selection import RFE
from sklearn.ensemble import GradientBoostingRegressor
import pandas as pd
import numpy as np

# Step 1: Define Targets
# Define targets based on the project description, such as income inequality metrics and other key socioeconomic indicators
targets = [
    '2018 Gini Coefficient_y', '2021 Gini Coefficient_y',
    '2023p Gini Coefficient_y', '2023p Palma Ratio_y'
]

# Define the features by excluding targets and other non-feature columns
non_features = ['Country', 'Region/Province', 'City/Town'] + targets
features = main_merged_dataset.drop(columns=non_features)

# Extract target variables
y = main_merged_dataset[targets]

# Exclude columns that are object types for feature selection
numerical_features = features.select_dtypes(include=['number'])

# Step 2: Variance Thresholding to Remove Low Variance Features
# Remove features with near-zero variance to eliminate features that provide little information
variance_threshold = 0.01  # Adjusted threshold based on domain knowledge
variance_selector = VarianceThreshold(threshold=variance_threshold)
features_var_reduced = variance_selector.fit_transform(numerical_features)
selected_features_var = numerical_features.columns[variance_selector.get_support(indices=True)]

# Create a DataFrame of the reduced features after variance thresholding
features_var_reduced_df = pd.DataFrame(features_var_reduced, columns=selected_features_var)
print("Features after variance thresholding shape:", features_var_reduced_df.shape)

# Step 3: Statistical Significance Test using SelectKBest (ANOVA F-statistic)
# Select top features that are statistically significant in predicting targets
k_best_selector = SelectKBest(score_func=f_regression, k='all')
k_best_selector.fit(features_var_reduced_df, y['2018 Gini Coefficient_y'])  # Example target column for scoring
stat_significant_features = features_var_reduced_df.columns[k_best_selector.get_support(indices=True)]

# Create a DataFrame of features after statistical significance test
features_stat_significant_df = features_var_reduced_df[stat_significant_features]
print("Features after statistical significance selection shape:", features_stat_significant_df.shape)

# Step 4: Recursive Feature Elimination (RFE)
# Using Gradient Boosting Regressor to rank feature importance and recursively eliminate weaker features
model = GradientBoostingRegressor()
rfe_selector = RFE(estimator=model, n_features_to_select=20)  # Keep the top 20 features
features_rfe_reduced = rfe_selector.fit_transform(features_stat_significant_df, y['2018 Gini Coefficient_y'])
selected_features_rfe = features_stat_significant_df.columns[rfe_selector.get_support(indices=True)]

# Create a DataFrame of features after RFE
features_final_df = pd.DataFrame(features_rfe_reduced, columns=selected_features_rfe)
print("Final selected features shape:", features_final_df.shape)

# Reasoning:
# - Variance Thresholding removes features with very low variance, which do not add useful information.
# - Statistical significance testing (using ANOVA F-statistic) ensures the features are relevant to predicting the target.
# - Recursive Feature Elimination (RFE) helps in selecting the most important features while eliminating redundant ones.
# - By using a combination of techniques, we can ensure our feature set is comprehensive, yet concise, optimizing model performance and reducing multicollinearity.
```

This comprehensive selection ensures our model will train on features that are relevant, improving accuracy and reducing overfitting.

```python
# Residual Analysis for Retrained Models
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Function to plot residuals and print residual analysis in text form
def residual_analysis(retrained_models_results_df, X_test, y_test):
    for index, row in retrained_models_results_df.iterrows():
        model = row['Best Estimator']
        target = row['Target']
        y_target_test = y_test[target]
        
        # Predict on the test set
        y_pred = model.predict(X_test)
        
        # Calculate residuals
        residuals = y_target_test - y_pred
        
        # Plot residuals
        plt.figure(figsize=(10, 6))
        sns.histplot(residuals, kde=True)
        plt.title(f'Residual Analysis for {target}')
        plt.xlabel('Residuals')
        plt.ylabel('Frequency')
        plt.show()
        
        # Print residual statistics
        print(f"\nResidual Analysis for {target}:")
        print(f"Mean Residual: {np.mean(residuals)}")
        print(f"Residual Standard Deviation: {np.std(residuals)}")
        print(f"Maximum Residual: {np.max(residuals)}")
        print(f"Minimum Residual: {np.min(residuals)}")

# Perform residual analysis on retrained models
residual_analysis(retrained_models_results_df, X_test_final, y_test)
```

```python
# Impact Assessment: Quantifying Relationships Between Variables
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to calculate and display impact assessment
# This function will quantify how population growth and climate patterns impact income inequality metrics
# and help understand the relationships for each target variable.
def impact_assessment(models_results_df, X_test, y_test):
    assessment_results = []
    
    for index, row in models_results_df.iterrows():
        model = row['Best Estimator']
        target = row['Target']
        y_target_test = y_test[target]
        y_pred = model.predict(X_test)
        
        # Calculate impact measures using differences between predictions and actuals
        residuals = y_target_test - y_pred
        mse = np.mean(residuals**2)
        rmse = np.sqrt(mse)
        
        # Impact on key metrics: consider impact in context of variance across the regions or deciles
        explained_variance = 1 - (np.var(residuals) / np.var(y_target_test))
        
        # Summarize the impact analysis
        assessment_results.append({
            'Target': target,
            'Mean Squared Error': mse,
            'Root Mean Squared Error': rmse,
            'Explained Variance': explained_variance
        })
        
        # Visualization of the actual vs predicted for impact assessment
        plt.figure(figsize=(10, 6))
        plt.scatter(y_target_test, y_pred, alpha=0.5)
        plt.plot([y_target_test.min(), y_target_test.max()], [y_target_test.min(), y_target_test.max()], 'k--', lw=2)
        plt.xlabel('Actual Values')
        plt.ylabel('Predicted Values')
        plt.title(f'Impact Assessment for {target} - Actual vs Predicted')
        plt.show()
    
    # Create DataFrame of assessment results
    impact_assessment_df = pd.DataFrame(assessment_results)
    print("\nImpact Assessment Results:")
    print(impact_assessment_df)

# Perform impact assessment on retrained models
impact_assessment(retrained_models_results_df, X_test_final, y_test)

# Conclusion:
# - The impact assessment results help determine how well our models have predicted income inequality metrics in the context of population growth and climate patterns.
# - We used RMSE and Explained Variance to quantify the accuracy and reliability of our models in analyzing and explaining these relationships.
# - The visualizations further support an understanding of how the model outputs relate to the actual data.
```
